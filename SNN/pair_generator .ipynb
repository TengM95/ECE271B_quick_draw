```python
from os import listdir
from os.path import isfile, join
mypath = '../data_little'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
with open('../label_new_little.txt') as f:
    lines = f.readlines();
    
```

```python
dicts = {}
for i in lines:
    i = i.strip('\n')
    i_split = i.split('|')
    dicts[i_split[1]] = [int(i_split[2]),int(i_split[0])]
```

```python
def pad_with(vector, pad_width, iaxis, kwargs):
    pad_value = kwargs.get('padder', 10)
    vector[:pad_width[0]] = pad_value
    vector[-pad_width[1]:] = pad_value
    return vector
```

```python
dirct = '../data_little/'
out_dict = '../data_combine/'
import numpy as np 
import os.path

for i in range(len(onlyfiles)):
    fn = onlyfiles[i]
    f_out = None;
    if not os.path.isfile(dirct + fn):
        continue
    print(dirct + fn)
    pos_data = np.load(dirct + fn)
    class_name_pos = fn.split('.')[0]
    #number = dicts[class_name_pos][1]
    #label = dicts[class_name_pos][0]
#     sample_index = np.random.choice(pos.shape[0], int(0.1*number))
#     sample = pos[sample_index]
#     del pos
#     pos_data = np.pad(pos, 1, pad_with,padder=label)[1:-1,1:]
#     print(pos_data.shape)

    for j in range(len(onlyfiles)):
        neg = onlyfiles[j]
        if not os.path.isfile(dirct +neg):
            continue
        if neg is fn:
            continue 
        else:
            print(dirct+neg)
            neg_data = np.load(dirct +neg)
            #class_name_neg = neg.split('.')[0]
            #number_neg = dicts[class_name_neg][1]
            #label_neg = dicts[class_name_neg][0]
#             neg_index = np.random.choice(neg_file.shape[0], int(0.1*number_neg))
#             sample_neg = neg_file[neg_index]
#             del neg_file
            #neg_data = np.pad(neg_file, 1, pad_with,padder=label_neg)[1:-1,1:]
#             print(neg_data.shape)
#             combination = np.meshgrid(pos_data,neg_data).T.reshape(-1,1570)
#             print(combination.shape)
            np.random.shuffle(neg_data)
            np.random.shuffle(pos_data)
            pos_length = pos_data.shape[0]
            neg_length = neg_data.shape[0]
            min_length = min(pos_length,neg_length)
            min_length = min(min_length,1000)
            pos_0 = pos_data[:min_length]
            np.random.shuffle(pos_data)
            pos_1 = pos_data[:min_length]
            neg_0 = neg_data[:min_length]
            combination = np.concatenate((pos_0,pos_1,neg_0),axis=1)
            if f_out is None:
                f_out = combination
            else:
                print(f_out.shape)
                print(combination.shape)
                f_out = np.vstack((f_out, combination))
    np.save(out_dict+class_name_pos+"combined", f_out)
```
