```python
f = open("categories.txt","r")
# And for reading use
classes = f.readlines()
f.close()
classes[-1] = classes[-1] + "\n"
```

```python
from six.moves import urllib
def download():
    base = 'https://storage.googleapis.com/quickdraw_dataset/full/numpy_bitmap/'
    for c in classes:
        cls_url = c.replace(' ', '%20')
        path = base+cls_url[0:-1]+'.npy'
        print(path)
        c_1 = c.replace(' ', '_')
        urllib.request.urlretrieve(path, 'data_all/'+c_1[0:-1]+'.npy')
```

```python
download()
```
